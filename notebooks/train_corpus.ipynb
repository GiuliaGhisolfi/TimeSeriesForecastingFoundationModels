```python
import os
from datasets import Dataset

def load_dataset_from_disk(dataset_name, path=f"{DATA_PATH}splitted_moirai_dataset"):
    save_path = f"{path}/{dataset_name.replace('/', '_')}"
    
    if os.path.exists(save_path):
        return Dataset.load_from_disk(save_path)
    else:
        raise FileNotFoundError(f"Dataset {dataset_name} not found in {save_path}")
```

```python
import pandas as pd
import numpy as np

def get_dataset_name(dataset):
    dataset_name = dataset.split("/")[-1]
    source = dataset.split("/")[0]
    return dataset_name, source

def get_num_variates_and_len(targets):
    num_variates_list = []
    ts_length_list = []

    for t in targets:
        arr = np.array(t)
        if arr.ndim == 1:
            ts_length_list.append(len(arr))
            num_variates_list.append(1)
        elif arr.ndim == 2:
            ts_length_list.append(arr.shape[0])
            num_variates_list.append(arr.shape[1])
        else:
            raise ValueError(f"Unexpected shape: {arr.shape}")
    
    return {
        "mean_num_variates": np.mean(num_variates_list),
        "min_num_variates": np.min(num_variates_list),
        "max_num_variates": np.max(num_variates_list),
        "mean_ts_length": np.mean(ts_length_list),
        "min_ts_length": np.min(ts_length_list),
        "max_ts_length": np.max(ts_length_list),
    }
```

```python
def get_info(dataset_path):
    ds = load_dataset_from_disk(dataset_path)

    dataset, source = get_dataset_name(str(ds.data["dataset"][0]))
    freq = str(ds.data["freq"][0])
    rows = len(ds.data["target"])

    stats = get_num_variates_and_len(ds.data["target"])

    print(f"{dataset_path} done :)")

    return {
        "dataset": dataset,
        "source": source,
        "freq": freq,
        "rows": rows,
        "num_variates": stats["mean_num_variates"],
        "min_num_variates": stats["min_num_variates"],
        "max_num_variates": stats["max_num_variates"],
        "ts_length": stats["mean_ts_length"],
        "min_ts_length": stats["min_ts_length"],
        "max_ts_length": stats["max_ts_length"],
    }
```
